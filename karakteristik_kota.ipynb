```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Membaca dataset dari file Excel
file_path = 'path_to_your_excel_file.xlsx'  # Ganti dengan path file Excel Anda
df = pd.read_excel(file_path)

# Memilih fitur untuk clustering
features = df[['Jumlah Penduduk (juta jiwa)', 'Luas Wilayah (km²)', 
                'Tingkat Pengangguran (%)', 'PDRB per Kapita (juta rupiah)']]

# Standarisasi data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Melakukan K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Menampilkan hasil clustering
print(df[['Kota', 'Cluster']])

# Visualisasi kluster
plt.figure(figsize=(10, 6))
plt.scatter(df['Jumlah Penduduk (juta jiwa)'], df['PDRB per Kapita (juta rupiah)'], 
            c=df['Cluster'], cmap='viridis')
plt.title('Clustering Kota-kota di Indonesia')
plt.xlabel('Jumlah Penduduk (juta jiwa)')
plt.ylabel('PDRB per Kapita (juta rupiah)')
plt.colorbar(label='Cluster')
plt.grid()
plt.show()

# Menghitung karakteristik masing-masing kluster
cluster_summary = df.groupby('Cluster').mean().reset_index()
print("\nRata-rata Karakteristik per Kluster:")
print(cluster_summary)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Atur variabel lingkungan untuk menghindari masalah memori
os.environ["OMP_NUM_THREADS"] = "1"

# Membaca data dari file Excel
df = pd.read_excel('kota_indonesia.xlsx')

# Mengambil kolom yang diperlukan untuk clustering
data = df[['Jumlah Penduduk (juta jiwa)', 'Luas Wilayah (km²)', 'Tingkat Pengangguran (%)', 'PDRB per Kapita (juta rupiah)']]

# Melakukan clustering dengan KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Kluster'] = kmeans.fit_predict(data)

# Menampilkan hasil klustering
print(df[['Kota', 'Kluster']])

# Menghitung rata-rata karakteristik setiap kluster
cluster_summary = df.groupby('Kluster').mean()
print("\nRata-rata karakteristik setiap kluster:")
print(cluster_summary)

# Visualisasi Kluster
plt.figure(figsize=(12, 6))

# Plot Jumlah Penduduk vs PDRB per Kapita
sns.scatterplot(data=df, x='Jumlah Penduduk (juta jiwa)', y='PDRB per Kapita (juta rupiah)',
                hue='Kluster', palette='viridis', style='Kluster', s=100)

# Menambahkan judul dan label
plt.title('Visualisasi Kluster Kota di Indonesia')
plt.xlabel('Jumlah Penduduk (juta jiwa)')
plt.ylabel('PDRB per Kapita (juta rupiah)')
plt.legend(title='Kluster')
plt.grid()

# Tampilkan plot
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os

# Atur variabel lingkungan untuk menghindari masalah memori
os.environ["OMP_NUM_THREADS"] = "1"

# Membaca data dari file Excel
df = pd.read_excel('karakteristik_kota.xlsx')

# Mengambil kolom yang diperlukan untuk clustering
data = df[['Jumlah Penduduk (juta jiwa)', 'Luas Wilayah (km²)', 'Tingkat Pengangguran (%)', 'PDRB per Kapita (juta rupiah)']]

# Melakukan clustering dengan KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Kluster'] = kmeans.fit_predict(data)

# Menampilkan hasil klustering
print(df[['Kota', 'Kluster']])

# Menghitung rata-rata karakteristik setiap kluster
numeric_cols = df.select_dtypes(include=['number']).columns
cluster_summary = df.groupby('Kluster')[numeric_cols].mean()
print("\nRata-rata karakteristik setiap kluster:")
print(cluster_summary)

# Visualisasi kluster
plt.figure(figsize=(10, 6))
plt.scatter(data['Jumlah Penduduk (juta jiwa)'], data['PDRB per Kapita (juta rupiah)'], c=df['Kluster'], cmap='viridis', marker='o')
plt.title('Visualisasi Kluster Kota Indonesia')
plt.xlabel('Jumlah Penduduk (juta jiwa)')
plt.ylabel('PDRB per Kapita (juta rupiah)')
plt.colorbar(label='Kluster')
plt.grid()
plt.show()
```
